```python
accountName = 'Francis'
accountBalance = 100
accountPassword = 'soup'

while True:
    print(f'{'#'*50}\n'
          f'Press b - balance\n'
          f'Press d - deposit\n'
          f'Press w - withdraw\n'
          f'Press s - show account\n'
          f'Press q - quit\n')

    # Input Processing
    action = input(f'What do you want to do?')
    action = action.lower()
    action = action[0]

    # Banking Commands
    commands = ['b','d','w','s','q']
    if action not in commands:
        print(f'Invalid command')
        break
    else:
        print(f'All within commands')
        if action == 'b':
            print(f'Get Balance')
            userPassword = input('Please enter the password: ')
            if userPassword != accountPassword:
                print('Incorrect password')
            else:
                print(f'Your balance is: {accountBalance}')
        elif action == 'q':
            print(f'End Session')
            break
```
